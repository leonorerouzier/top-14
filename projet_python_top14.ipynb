```python
import pandas as pd
import datetime as dt
data = pd.read_csv('Data_projetTOP14.csv')

# Selecting all data from the 2021-2022 season
last_season = data[data['season'] == '2021-2022']
last_season
```

```python
import numpy as np
# Setting up a dictionnary of all players from the 2021-2022 season
players_dict = {}
seen_players = set()

# Iterate through each row in last_season DataFrame
for index, row in last_season.iterrows():
    home_team = row['home_team']
    away_team = row['away_team']

    # Extract home and away team players
    home_team_players = set(row['home_1':'home_23'].values)
    away_team_players = set(row['away_1':'away_23'].values)

    # Update players_dict with unique home and away team players
    if home_team not in players_dict:
        players_dict[home_team] = home_team_players
    else:
        players_dict[home_team].update(home_team_players)

    if away_team not in players_dict:
        players_dict[away_team] = away_team_players
    else:
        players_dict[away_team].update(away_team_players)

    # Update the set of seen players
    seen_players.update(home_team_players)
    seen_players.update(away_team_players)

# Pad player lists with NaN to ensure all arrays have the same length
max_players = 52

for team, players in players_dict.items():
    players_dict[team] = list(players) + [np.nan] * (max_players - len(players))

pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', None)

# Create the pandas DataFrame
last_season_players = pd.DataFrame(players_dict)

# Print dataframe
print(last_season_players)
```

```python
# Compare 'home_score' and 'away_score' to determine the result for each game
def determine_result(row):
    if row['home_score'] > row['away_score']:
        return "win"
    elif row['home_score'] < row['away_score']:
        return "loss"
    else:
        return "draw"

# Create a new DataFrame to store teams, day, and their results
team_results = pd.DataFrame(columns=['Team', 'day', 'Result'])

# Add data from 'home_team', 'day', and 'Result' for home team
home_team_results = last_season[['home_team', 'day', 'home_score', 'away_score']].rename(columns={'home_team': 'Team'})
home_team_results['Result'] = home_team_results.apply(determine_result, axis=1)

# Add data from 'away_team', 'day', and 'Result' for away team
away_team_results = last_season[['away_team', 'day', 'home_score', 'away_score']].rename(columns={'away_team': 'Team'})
away_team_results['Result'] = away_team_results.apply(determine_result, axis=1)

# Concatenate the results for home and away teams
team_results = pd.concat([home_team_results, away_team_results], ignore_index=True)

# Drop unnecessary columns
team_results.drop(['home_score', 'away_score'], axis=1, inplace=True)

def display_team_results():

    # Create a dictionary to store the results for each day
    results = {}

    # Loop through each unique day in the team_results DataFrame and create a tab for each day
    for day in team_results['day'].unique():
        # Filter the team_results DataFrame for the current day
        day_results = team_results[team_results['day'] == day]

        # Create a tab for the day containing team results
        day_tab = day_results.to_string(index=False)

        # Add the tab to the results dictionary
        results[day] = day_tab

    # Ask the user to input a day between J1 and J26
    user_input = input("Type a day between 'J1' and 'J26': ")

    # Check the user's input and display the tab for the corresponding day
    if user_input in results:
        print(results[user_input])
    else:
        print("Invalid input. Please type a day between 'J1' and 'J26'.")
```

```python
def display_cumulative_ranking():

    # Define a mapping of 'day' values to their numeric representation
    day_mapping = {'J1': 1, 'J2': 2, 'J3': 3, 'J4': 4, 'J5': 5, 'J6': 6, 'J7': 7, 'J8': 8, 'J9': 9, 'J10': 10,
               'J11': 11, 'J12': 12, 'J13': 13, 'J14': 14, 'J15': 15, 'J16': 16, 'J17': 17, 'J18': 18, 'J19': 19,
               'J20': 20, 'J21': 21, 'J22': 22, 'J23': 23, 'J24': 24, 'J25': 25, 'J26': 26}

    # Ask the user to input a day between J1 and J26
    user_input = input("Type a day between 'J1' and 'J26': ")

    # Check the user's input and display the data from J1 to the specified day
    if user_input in day_mapping:
        matchday = day_mapping[user_input]
    
        # Filter the data based on the numeric representation
        selected_data = team_results[team_results['day'].map(day_mapping) <= matchday]

        # Create a dictionary to map results to points
        points_mapping = {'win': 4, 'draw': 2, 'loss': 0}

        # Map the 'Result' column to points for the selected data
        selected_data['Points'] = selected_data['Result'].map(points_mapping)

        # Initialize a list to store dictionaries for cumulative ranking
        cumulative_ranking_list = []

        # Group the selected data by 'Team' and calculate cumulative points
        grouped = selected_data.groupby('Team')

        for team, group in grouped:
            wins = group['Result'].eq('win').sum()
            losses = group['Result'].eq('loss').sum()
            draws = group['Result'].eq('draw').sum()
            cumulative_points = wins * 4 + draws * 2

            team_data = {
            'Team': team,
            'Win': wins,
            'Loss': losses,
            'Draw': draws,
            'Cumulative_Points': cumulative_points
            }

            cumulative_ranking_list.append(team_data)

        # Create the DataFrame from the list of dictionaries
        cumulative_ranking_df = pd.DataFrame(cumulative_ranking_list)

        # Sort the cumulative ranking by 'Cumulative_Points' in descending order
        cumulative_ranking_df = cumulative_ranking_df.sort_values(by='Cumulative_Points', ascending=False)

        # Display the resulting filtered ranking DataFrame
        print(f"TOP14 Ranking as of {user_input}:")
        print(cumulative_ranking_df)
    else:
        print("Invalid input format. Please enter a day between 'J1' and 'J26'.")
```

```python
# Impact on home vs away games for Teams (over the whole season)

def display_home_away_impact():
    # Define the cutoff date
    cutoff_date = '2022-06-05'

    # Filter the DataFrame to keep only the games before or on the cutoff date
    filtered_df = last_season[last_season['date'] <= cutoff_date]

    # Now, filtered_df contains only the games that happened on or before 2022-06-05
    # Calculate the % of home wins for each team
    home_wins = filtered_df[filtered_df['home_score'] > filtered_df['away_score']]
    home_win_counts = home_wins['home_team'].value_counts()
    total_home_games = filtered_df['home_team'].value_counts()
    home_win_percent = (home_win_counts / total_home_games) * 100

    # Calculate the % of away wins for each team
    away_wins = filtered_df[filtered_df['away_score'] > filtered_df['home_score']]
    away_win_counts = away_wins['away_team'].value_counts()
    total_away_games = filtered_df['away_team'].value_counts()
    away_win_percent = (away_win_counts / total_away_games) * 100

    # Calculate the overall percentage of wins for each team
    overall_win_percent = (home_win_counts + away_win_counts) / (total_home_games + total_away_games) * 100

    # Create a DataFrame to store the results
    results_percentage = pd.DataFrame({
    'Team': home_win_percent.index,
    '% of home wins': home_win_percent.values,
    '% of away wins': away_win_percent.values,
    'Overall % of wins': overall_win_percent.values}).set_index('Team')  # Set 'Team' column as the index

    # Get the list of available columns for ranking
    available_columns = ['% of home wins', '% of away wins', 'Overall % of wins']

    # Prompt the user to choose a column for ranking
    print("Available columns for ranking:", available_columns)
    chosen_column = input("Enter the column name for ranking: ")

    # Check if the chosen column is valid
    if chosen_column in available_columns:
        # Rank the teams based on the chosen column in descending order (best to worst)
        results_percentage['Rank'] = results_percentage[chosen_column].rank(ascending=False)

        # Sort the DataFrame based on the rank
        results_percentage.sort_values(by='Rank', inplace=True)

        # Display the resulting DataFrame
        print(results_percentage)
    else:
        print("Invalid column name. Please choose a valid column for ranking.")
```
